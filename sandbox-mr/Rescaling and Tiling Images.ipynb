```python
import numpy as np
import os
from pathlib import Path
from PIL import Image, ImageDraw, ImageFont
```

```python
im = Image.new('RGBA', (2048, 2048), color='#000')

fnt = ImageFont.truetype("arial.ttf", 20)

d = ImageDraw.Draw(im)
```

```python
n = 16
spacing = 2048 / n

for i in range(n):
    for j in range(n):
        d.text(
            (spacing * (i + 1) - spacing / 2, spacing * (j + 1) - spacing / 2),
            "(%02d, %02d)" % (i, j),
            font=fnt,
            fill="#FFF",
            anchor="mm",
        )
```
