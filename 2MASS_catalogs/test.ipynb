```python
import astropy
from astropy.io import fits
from astropy.table import Table
import pandas
import numpy as np
import matplotlib.pyplot as plt

c = astropy.constants.c.to("km/s")
```

```python
# open the file
# convert to table then to pandas
# rename ra and dec columns
# filter to only have highest redshift (0.08 - 0.09)
# write into new file 
```

```python
fits_table_filename = 'catalog/2mrs_1175_done.fits'

hdul = fits.open(fits_table_filename)
data = Table(hdul[1].data) # assuming the first extension is a table
df = data.to_pandas()
```

```python
z_cond =  np.logical_and(0.08 <= df["V"]/c, df["V"]/c <= 0.09)
```

```python
df2 = df[z_cond]
```

```python
df2 = df2.rename(columns={"TMID": "id", "RA": "ra", "DEC": "dec"})
```
